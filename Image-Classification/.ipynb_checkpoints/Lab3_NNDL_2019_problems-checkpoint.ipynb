```python
import numpy as np
```

```python
from scipy import signal
f = [1,-1,-2,0,-1,1,2,-1]
h = [1,2,-2]
con_same =  signal.convolve(f,h, mode = 'same')
print('same = ', con_same)
con_full =  signal.convolve(f,h, mode = 'full')
print('full = ', con_full)
con_valid =  signal.convolve(f,h, mode = 'valid')
print('valid = ', con_valid)

#Full = N+M-1
#same = Max(M,N)
#valid = Max(M,N) - Min(M,N) +1
```

```python
from matplotlib import pyplot as plt
def visualize(img):
    plt.imshow(img,'gray')
    plt.colorbar()
    plt.show()
    print('Image size:', img.shape)
```

```python
from scipy import signal

h = np.array([[1,1],[1,1]])
for i  in range(0,5):
     h = signal.convolve2d(h,h)

        
visualize(h)

print(h)
```

```python
from matplotlib import pyplot as plt
from matplotlib.image import imread
img = imread('MR15044.jpeg')
visualize(img)

RGB_sum = np.sum(img,axis = 2)
visualize(RGB_sum)

img_valid = signal.convolve2d(RGB_sum,h, mode = 'valid')
visualize(img_valid)
img_same = signal.convolve2d(RGB_sum, h, mode = 'same')
visualize(img_same)
```

```python
visualize(RGB_sum[2::5,2::5])

visualize(img_same[2::5,2::5])
```

```python
import tensorflow as tf
#import tensorflow.contrib.eager as tfe
import keras as k

print('TenserFlow version '+tf.__version__)
tf.enable_eager_execution()
if tf.executing_eagerly():
    print('Eager execution activated!')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D
f = np.array([1,-1,-2,0,-1,1,2,-1])
f = f.reshape(1,8,1)
h = np.array([-2,2,1])
h = h.reshape(3,1,1)
model = Sequential()
model.add(Conv1D(1,3, strides = 1, padding = 'same',input_shape = (8,1)))
weights = model.layers[0].get_weights()
weights[0] = h

model.layers[0].set_weights(weights)
output = model.predict(f)    
print('same: ',output.reshape(8,))
```
